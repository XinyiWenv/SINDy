```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "active_power_total.csv"  # Update the path if necessary
data = pd.read_csv(file_path)

# Check the structure of the data
print(data.head())

# Convert 'Timestamp' column to datetime if not already in datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(data['Timestamp'][0:3600], data['Active Power Total (Mean)'][0:3600], label='Active Power Total', color='blue')

# Add titles and labels
plt.title('Active Power Total Over Time', fontsize=16)
plt.xlabel('Timestamp', fontsize=14)
plt.ylabel('Active Power Total (Mean)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
